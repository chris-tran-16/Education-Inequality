```python
data_merged3 = data_merged2.copy()
data_merged3.loc[data_merged3['GSLO'] == 'N', ('GSLO')] = -3
data_merged3.loc[data_merged3['GSLO'] == 'UG', ('GSLO')] = -2
data_merged3.loc[data_merged3['GSLO'] == 'PK', ('GSLO')] = -1
data_merged3.loc[data_merged3['GSLO'] == 'KG', ('GSLO')] = 0

data_merged3.loc[data_merged3['GSHI'] == 'N', ('GSLO')] = -3
data_merged3.loc[data_merged3['GSHI'] == 'UG', ('GSLO')] = -2
data_merged3.loc[data_merged3['GSHI'] == 'PK', ('GSLO')] = -1
data_merged3.loc[data_merged3['GSHI'] == 'KG', ('GSLO')] = 0


data_merged3 = data_merged3.drop(data_merged3['GSLO'] == -2)

ungrade = ((data_merged3['GSLO'] == -2) | (data_merged3['GSLO'] == -3)
 | (data_merged3['GSHI'] == -2) | (data_merged3['GSHI'] == -3))

data_merged3 = data_merged3[~ungrade]
```
